```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'anime.csv')
df.head()
```

```python
df.loc[1]['Title']
```

```python
def extract_title(txt):
    check = False
    data = ''
    
    for i in txt:
        if i == ')':
            check = False
            break
        if check == True:
            data += i
        if i == '(':
            check = True
        
       
    return data
```

```python
df['Title'].apply(extract_title)
```
